```python
%matplotlib inline
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import load_wine
from sklearn.model_selection import cross_val_score
import matplotlib.pyplot as plt
```

```python
wine = load_wine()
```

```python
from sklearn.model_selection import train_test_split
Xtrain, Xtest, Ytrain, Ytest = train_test_split(wine.data,wine.target,test_size=0.3)
clf = DecisionTreeClassifier(random_state=0)
rfc = RandomForestClassifier(random_state=0)

clf = clf.fit(Xtrain,Ytrain)
rfc = rfc.fit(Xtrain,Ytrain)

score_c = clf.score(Xtest,Ytest)
score_r = rfc.score(Xtest,Ytest)
```

```python
print("single tree:{}".format(score_c),"ramdon forest:{}".format(score_r))
```

```python
#交叉验证
rfc = RandomForestClassifier(n_estimators=25) #建立25棵决策树
rfc_s = cross_val_score(rfc,wine.data,wine.target,cv=10) #数据分10份,共10次

clf = DecisionTreeClassifier()
clf_s = cross_val_score(clf,wine.data,wine.target,cv=10)
clf_s
```

```python
plt.plot(range(1,11),rfc_s,label = "RandonForest")
plt.plot(range(1,11),clf_s,label = "Decision Tree")
plt.legend()
plt.show()
```

```python
superpa = []
for i in range(200):
    rfc = RandomForestClassifier(n_estimators=i+1,n_jobs=-1) #设置森林中多少棵树
    rfc_s = cross_val_score(rfc,wine.data,wine.target,cv=10).mean()
    superpa.append(rfc_s)
print(max(superpa),superpa.index(max(superpa))) #list.index(object)返回object在list中的索引
plt.figure(figsize=[20,5])
plt.plot(range(1,201),superpa)
plt.show()
```
