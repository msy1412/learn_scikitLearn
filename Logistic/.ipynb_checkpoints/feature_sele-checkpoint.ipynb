```python
from sklearn.linear_model import LogisticRegression as lr
from sklearn.datasets import load_breast_cancer
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn.feature_selection import SelectFromModel 
import matplotlib.pyplot as plt
```

```python
data = load_breast_cancer()
```

```python
x = data.data
y = data.target
data.data.shape
```

```python
LR_=lr(solver="liblinear",C=0.8,random_state=420)
cross_val_score(LR_,data.data,data.target,cv=10).mean()
```

```python
x_embeded = SelectFromModel(LR_,norm_order=1).fit_transform(data.data,data.target)
x_embeded.shape #x_embeded是去掉无用特征的矩阵
```

```python
cross_val_score(LR_,x_embeded,data.target,cv=10).mean()
```

```python
fullx=[]
fsx=[]
```

```python
threshold = np.linspace(0,abs(LR_.fit(data.data,data.target).coef_).max(),20)
```

```python
k=0
for i in threshold:  #随着阈值的提高,特征越来越少
    x_embeded = SelectFromModel(LR_,threshold=i).fit_transform(data.data,data.target)
    fullx.append(cross_val_score(LR_,data.data,data.target,cv=5).mean())
    fsx.append(cross_val_score(LR_,x_embeded,data.target,cv=5).mean())
    print((threshold[k],x_embeded.shape))
    k+=1
```

```python
plt.figure(figsize=(20,5))
plt.plot(threshold,fullx,label="full")
plt.plot(threshold,fsx,label="feature selection")
plt.xticks(threshold)
plt.legend()
plt.show()
```

```python
full=[]
fs=[]
C = np.arange(5.50,6.51,0.01)
```

```python
for i in C:   #训练C值
    LR = lr(solver="liblinear",C=i,random_state=420)
    full.append(cross_val_score(LR,data.data,data.target,cv=10).mean())
    
    x_embeded = SelectFromModel(LR,norm_order=1).fit_transform(data.data,data.target)
    fs.append(cross_val_score(LR,x_embeded,data.target,cv=10).mean())
    
    #print(max(fs),C[fs.index(max(fs))]) #每次都打出最大值
    
print(max(fs),C[fs.index(max(fs))])  #最后才打出最大值
```

```python
plt.figure(figsize=(20,5))
plt.plot(C,full,label="full")
plt.plot(C,fs,label="feature sele")
plt.xticks(C)
plt.legend()
plt.show()
```

```python
LR_=lr(solver="liblinear",C=5.87,random_state=420)  #使用经过学习后的C值=5.87
cross_val_score(LR_,data.data,data.target,cv=10).mean()
```

```python
LR_=lr(solver="liblinear",C=5.87,random_state=420)  #使用经过学习后的C值=5.87
x_embeded = SelectFromModel(LR_,norm_order=1).fit_transform(data.data,data.target)  #再进行特征选择
cross_val_score(LR_,x_embeded,data.target,cv=10).mean()
```
