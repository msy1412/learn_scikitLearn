```python
from sklearn.datasets import load_breast_cancer
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import numpy as np
import pandas as  pd
from time import time
import datetime
```

```python
data = load_breast_cancer()
x = data.data
y = data.target
```

```python
x.shape
```

```python
y.shape
```

```python
np.unique(y) #y的值的种类
```

```python
plt.scatter(x[:,0],x[:,1],c=y)
plt.show()
```

```python
Xtrain, Xtest, Ytrain, Ytest = train_test_split(x,y,test_size=0.3,random_state=420)
```

```python
kernel = ["linear","rbf","sigmoid"] #degree = 1时 ,"poly"不能得出结果
for kernel in kernel :
    time0 = time()
    clf = SVC(kernel = kernel
              ,gamma="auto"
              #,degree = 1
              ,cache_size=3000 #内存允许占用
             ).fit(Xtrain,Ytrain)
    print("accuracy with %s is %s"%(kernel,clf.score(Xtest,Ytest)))
    print(datetime.datetime.fromtimestamp(time()-time0).strftime("%M:%S:%f"))
```

```python
kernel = ["linear","poly","rbf","sigmoid"] #degree = 1
for kernel in kernel :
    time0 = time()
    clf = SVC(kernel = kernel
              ,gamma="auto"
              ,degree = 1
              ,cache_size=3000 #内存允许占用
             ).fit(Xtrain,Ytrain)
    print("accuracy with %s is %s"%(kernel,clf.score(Xtest,Ytest)))
    print(datetime.datetime.fromtimestamp(time()-time0).strftime("%M:%S:%f"))
```

```python
from sklearn.preprocessing import StandardScaler

data = pd.DataFrame(x)
data.describe([0.01,0.05,0.1,0.25,0.5,0.75,0.9,0.99]).T
#有偏态问题,非正太分布
```

```python
#归一化
X = StandardScaler().fit_transform(x)
data = pd.DataFrame(X)
data.describe([0.01,0.05,0.1,0.25,0.5,0.75,0.9,0.99]).T
```

```python
#归一化后再训练的效果

Xtrain, Xtest, Ytrain, Ytest = train_test_split(X,y,test_size=0.3,random_state=420)

kernel = ["linear","poly","rbf","sigmoid"] #degree = 1
for kernel in kernel :
    time0 = time()
    clf = SVC(kernel = kernel
              ,gamma="auto"
              ,degree = 1
              ,cache_size=3000 #内存允许占用
             ).fit(Xtrain,Ytrain)
    print("accuracy with %s is %s"%(kernel,clf.score(Xtest,Ytest)))
    print(datetime.datetime.fromtimestamp(time()-time0).strftime("%M:%S:%f"))
```

```python
score = []
gamma_range = np.logspace(-10, 1, 50) #返回在对数刻度上均匀间隔的数字
for i in gamma_range:
    clf = SVC(kernel="rbf",gamma = i,cache_size=5000).fit(Xtrain,Ytrain)
    score.append(clf.score(Xtest,Ytest))
    
print(max(score), gamma_range[score.index(max(score))])
```

```python
plt.plot(gamma_range,score)
plt.show()
```

```python
from sklearn.model_selection import StratifiedShuffleSplit
from sklearn.model_selection import GridSearchCV
time0 = time()
```

```python
#网格搜索中输入输入到模型的参数
gamma_range = np.logspace(-10,1,20)
coef0_range = np.linspace(0,5,10)

param_grid = dict(gamma = gamma_range,coef0 = coef0_range)
cv = StratifiedShuffleSplit(n_splits=5, test_size=0.3, random_state=420)
```

```python
gamma_range
```

```python
coef0_range
```

```python
grid = GridSearchCV(SVC(kernel = "poly",degree=1,cache_size=5000)
                    ,param_grid = param_grid
                    ,cv = cv)
grid.fit(X,y)
print("The best parameters are %s with a score of %0.5f" % (grid.best_params_,grid.best_score_))
print(datetime.datetime.fromtimestamp(time()-time0).strftime("%M:%S:%f"))
```
